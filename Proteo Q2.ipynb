## MS/MS Spectra

Construct a plot the MS/MS spectrum (peptide sequence) for the peptide Ala-Ser-Gly-Leu-Phe-Asp-Glu-Arg.  On the graph, plot abundance of ions (y-axis) versus m/z on the x-axis.  *Assume the abundances are the same for all peptide fragment ions* and you only need to plot the y-ions.  Make sure to show the masses for each y-ion.

Ala-Ser-Gly-Leu-Phe-Asp-Glu-Arg

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Monotopic mass used
mass = {'Ala':71.04,'Ser':87.03,'Gly':57.02,'Leu':113.08,
        'Phe':147.07,'Asp':115.03,'Glu':129.04,'Arg':156.10}
water = 18.0
mass.update({'Arg':mass['Arg'] + water})

n = len(mass)-1
abundance = [9]*n
mass_values = list(mass.values())
y_ions = {}
for y in range(len(mass)-1,0,-1):
    y_ions['y'+str(y)] = np.sum(mass_values)
    mass_values.pop()
    
    
print(y_ions)
ion_list = list(y_ions.values())
  
# plot
fig, ax = plt.subplots()

ax.scatter(ion_list, abundance, color='red')
ax.vlines(ion_list,[0]*n,[9]*n, color='red')

# Fit labels on graph
for i, txt in enumerate(ion_list):
    if i == 6:
        ax.annotate(txt, (ion_list[i]+0.1, abundance[i]-0.55))
    elif i == 0:
        ax.annotate(txt, (ion_list[i]-50, abundance[i]+0.25))
    else:
        ax.annotate(txt, (ion_list[i], abundance[i]+0.25))
    
ax.set(ylim=(0, 11), yticks=np.arange(1, 11))
ax.set_xlabel('m/z')
ax.set_ylabel('Abundance')
ax.set_title('MS/MS Spectrum')
plt.show()
```
